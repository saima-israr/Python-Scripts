```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import random
from datetime import *

data = [['Inside Sales','Asia', '1/1/2020'], ['Inside Sales','Africa', '1/1/2020'], ['Lead Generation','Oceania', '1/7/2020'], ['Sales Support','Europe', '1/7/2020'], ['Outside Sales','Asia', '1/14/2020'], ['Sales Support','Europe', '1/14/2020'], ['Lead Generation','America', '1/21/2020'], ['Sales Support','Oceania', '1/21/2020'], ['Outside Sales','Oceania', '1/28/2020']]
df = pd.DataFrame(data,columns=['SalesType','Region','Date'])
print(df)

scatter_x = np.array(df['SalesType'])
scatter_y = np.array(pd.to_datetime(df['Date']))
group = np.array(df['Region'])

fig, ax = plt.subplots()
i = 0
for g in np.unique(group):
    ix = np.where(group == g)
    rgb = np.random.seed(3)
    ax.scatter(scatter_x[ix], scatter_y[ix], c = rgb, label = g, s = 200, marker = i+4)
    i=i+1              

ax.set_xlabel('SalesType', family='Arial', fontsize=9)
ax.set_ylabel('Date', family='Arial', fontsize=9)
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))

plt.title('Customer Data', family='Arial', fontsize=12)
plt.show()
```
