```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

data = [['Sale1','Asia', '1/1/2020'], ['Sale2','Africa', '1/7/2020'], ['Sale1','Asia', '1/14/2020'], ['Sale2','Africa', '1/14/2020'], ['Sale3','Europe', '1/14/2020'], ['Sale4','America', '1/21/2020'], ['Sale5','Oceania', '1/21/2020'], ['Sale5','Oceania', '1/28/2020']]
df = pd.DataFrame(data,columns=['SalesType','Region','Date'])
print(df)

color_dict = dict({'Africa':'brown',
                  'Asia':'green',
                  'Europe': 'orange',
                  'Oceania': 'red',
                   'America': 'dodgerblue'})

fig = plt.figure()
ax = fig.add_subplot()

sns.scatterplot(x="SalesType", y="Date", hue="Region",
              data=df, palette=color_dict, 
                   legend='full')
    
ax.set_xlabel('SalesType', family='Arial', fontsize=9)
ax.set_ylabel('Date', family='Arial', fontsize=9)
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.title('Customer Data', family='Arial', fontsize=12)
plt.show()
```
