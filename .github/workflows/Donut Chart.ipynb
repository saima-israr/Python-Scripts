```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import random


data = [['Asia', 'Africa', 'Europe', 'America', 'Oceania'], [100, 200, 300, 400, 500]]
rows = zip(data[0], data[1])
headers = ['Region', 'Sales']
df = pd.DataFrame(rows, columns=headers)

plt.pie(df['Sales'], labels=df['Region'], colors=np.random.seed(3), autopct='%1.1f%%', shadow=False)    
centre_circle = plt.Circle((0,0),0.45,color='grey', fc='white',linewidth=1.25)

fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.axis('equal')
plt.show() 
```
